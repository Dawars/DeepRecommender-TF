```python
import xmltodict
import codecs
import collections
```

```python
with codecs.open('./anime.xml', 'r', 'utf-8', errors='ignore') as f:
    anime_xml = f.read()
    
anime_xml[:1000]
```

```python
# parse xml file
anime_list = xmltodict.parse(anime_xml)['anime_list']['anime']
print('Number of anime', len(anime_list))
```

```python
#counting duplicates
dups = collections.defaultdict(int)

for anime in anime_list:
    dups[anime.get('title')] += 1
    
sorted(dups.items(), key=lambda x:x[1], reverse=True)
```

```python
# display data
keys_to_read = [
    'title',
    'episode_count',
    #'genres',
    'score',
    'popularity',
    #'status',
    'date_start'
]
for anime in anime_list:
    for key in keys_to_read:
        print(anime.get(key), '\t', end='')
    print()
```
