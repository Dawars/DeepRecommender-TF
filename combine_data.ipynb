```python
# remove spec chars and make lowercase BEFORE sorting

mal_titles['title'] = mal_titles.apply(lambda row: ''.join(e for e in row['title'].lower() if e.isalnum()), axis=1)
mangaki_titles['title'] = mangaki_titles.apply(lambda row: ''.join(e for e in row['title'].lower() if e.isalnum()), axis=1)
```

```python
mal_sorted = mal_titles.sort_values('title').reset_index()
mangaki_sorted = mangaki_titles.sort_values('title').reset_index()

display(mangaki_sorted.head(20))
display(mal_sorted.head(20))
```

```python
def get_item(df, i):
    return df.loc[i, 'title']
```

```python
matching_title = 0
i = j = 0

while i < len(mal_sorted) and j < len(mangaki_sorted):
    mal = get_item(mal_sorted, i)
    mangaki = get_item(mangaki_sorted, j)
    
    print('mal', i, mal)
    print('mangaki', j, mangaki)
    if mal == mangaki:
        
        mal_sorted.loc[i,'mangaki_id']= mangaki_sorted.loc[j, 'work_id']
        mangaki_sorted.loc[i,'mal_id']= mal_sorted.loc[j, 'anime_id']
        
        j += 1
        i += 1
        print('>', i, j, mal)
        matching_title += 1
    else:
        while get_item(mal_sorted, i) < get_item(mangaki_sorted, j):
            i += 1
        while get_item(mal_sorted, i) > get_item(mangaki_sorted, j):
            j += 1
```
